```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import deque
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, DBSCAN
from sklearn.model_selection import train_test_split
from Clustering import UGaussianMixture, UKMean, UDBSCAN
```

```python
data = pd.read_csv('DATA/iris.csv')
```

```python
X = data.drop('species', axis=1)
y = data['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
eps, min_samp = 0.5, 7

umodel = UDBSCAN(eps=eps, min_samples=min_samp)
umodel.fit(X)

model = DBSCAN(eps=eps, min_samples=min_samp)
model.fit(X)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_figwidth(13)
fig.set_dpi(200)
ax1.set_title('Реализованная модель')
ax2.set_title('Пакетная модель')
sns.scatterplot(X, x='sepal_length', y='petal_length', ax=ax1, hue=umodel.clusters_, palette='tab10')
sns.scatterplot(X, x='sepal_length', y='petal_length', ax=ax2, hue=model.labels_, palette='tab10');
```
