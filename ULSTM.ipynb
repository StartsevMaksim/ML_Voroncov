```python
class ULSTM:
    def __init__(self,
                 input_size,
                 hidden_size,
                 bias=True):
        self.input_size = input_size
        self.hidden_size = hidden_size
        #Слой забывания
        self.forget_layer_ = nn.USequential()
        self.forget_layer_.add_module('forget', nn.ULinear(input_size+hidden_size, hidden_size))
        self.forget_layer_.add_module('sigmoid', USigmoid())
        #Слой обновления контекста
        self.update_layer_ = nn.USequential()
        self.update_layer_.add_module('update', nn.ULinear(input_size+hidden_size, hidden_size))
        self.update_layer_.add_module('sigmoid', USigmoid())
        #Слой запоминания
        self.input_layer_ = nn.USequential()
        self.input_layer_.add_module('input', nn.ULinear(input_size+hidden_size, hidden_size))
        self.input_layer_.add_module('tanh', nn.UTanh())
        #Выходной слой
        self.output_layer_ = nn.USequential()
        self.output_layer_.add_module('output', nn.ULinear(input_size+hidden_size, hidden_size))
        self.output_layer_.add_module('sigmoid', USigmoid())
        
    def __repr__(self):
        return (f'LSTM(input_size={self.input_size}, '
                     f'hidden_size={self.hidden_size})')       

    def forward(self, X, h, c):
        x = torch.cat([h, X])
        forget = self.forget_layer_.forward(x)
        update = self.update_layer_.forward(x)
        input = self.input_layer_.forward(x)
        output = self.output_layer_.forward(x)
        c_forward = forget * c + update * input
        h_forward = output * torch.tanh(c_forward)
        return h_forward, h_forward, c_forward
```
